```python
df['DATE_DIED'] = df['DATE_DIED'].astype(str)

df.loc[df['DATE_DIED'] != '9999-99-99', 'DATE_DIED'] = '1'

df.head(100)
```

```python
print(df['DATE_DIED'].dtype)
```

```python
print(df['DATE_DIED'].value_counts())
```

```python
print(df['DATE_DIED'].unique())
df['DATE_DIED'] = df['DATE_DIED'].astype(str)
df['DATE_DIED'] = df['DATE_DIED'].replace('9999-99-99', '0')
df.loc[df['DATE_DIED'] != '0', 'DATE_DIED'] = '1'
df.head(100)
```

```python
df = df.drop('PATIENT_TYPE', axis=1)
df = df.drop('COPD', axis=1)
df = df.drop('ASTHMA', axis=1)
df = df.drop('INTUBED', axis=1)
df = df.drop('MEDICAL_UNIT', axis=1)
df = df.drop('USMER', axis=1)
df = df.drop('PREGNANT', axis=1)
```

```python
df.head()
```

```python
df = df.loc[~df['CLASIFFICATION_FINAL'].isin([4, 5, 6])]
df.head(100)
```

```python
# tạo list các giá trị cần giữ lại
keep_values = [1, 2, 3]

# xóa tất cả các hàng có giá trị kết quả khác 1, 2, 3
df = df[df['CLASIFFICATION_FINAL'].isin(keep_values)]
df = df.drop('CLASIFFICATION_FINAL', axis=1)
```

```python
#tạo list các giá trị cần giữ lại
keep_values = [1, 2]

# xóa tất cả các hàng có giá trị kết quả khác 1, 2
df = df[df['ICU'].isin(keep_values)]
```

```python
## Tạo list các giá trị cần giữ lại
keep_values = [1, 2]

# Xóa các hàng có giá trị không nằm trong danh sách giá trị cần giữ lại cho cột 
df = df[(df['HIPERTENSION'].isin(keep_values)) & (df['CARDIOVASCULAR'].isin(keep_values)) 
& (df['DIABETES'].isin(keep_values)) & (df['OBESITY'].isin(keep_values)) & (df['OTHER_DISEASE'].isin(keep_values))
& (df['TOBACCO'].isin(keep_values)) & (df['INMSUPR'].isin(keep_values)) & (df['RENAL_CHRONIC'].isin(keep_values))]
```

```python
# Lấy cột "kết quả" ra khỏi DataFrame
ket_qua = df.pop('DATE_DIED')

# Chèn cột "kết quả" vào cuối DataFrame
df.insert(len(df.columns), 'DATE_DIED', ket_qua)
df.head()
```

```python
# sắp xếp lại index
df = df.reset_index(drop=True)
```

```python
df.rename(columns={'DATE_DIED': 'RESULT'}, inplace=True)
```

```python
plt.figure(figsize=(13,11))
sns.heatmap(df.corr(),annot=True, fmt=".2f")
plt.title("Độ tương quan giữa các thuộc tính",fontsize=18)
```

```python
#Đồ thị phân bố giới tính
pie_columns = ["SEX"]
for column in pie_columns:
    plt.figure(figsize=(6, 6))  # Đặt kích thước của biểu đồ (4 inches x 4 inches)
    df[column].value_counts().plot(kind="pie", autopct='%1.1f%%')
    plt.title(column)
    plt.ylabel("")
    plt.show()
```

```python
# Đồ thị cột (bar chart) cho cột "AGE"
df['AGE'].plot(kind='hist', bins=10, xlabel='AGE', ylabel='Frequency', title='Distribution of AGE')
plt.show()
```

```python
#Biểu đồ đếm số lượng cột RENAL_CHRONIC và PNEUMONIA
fig, axs = plt.subplots(ncols=2, figsize=(15, 7))
sns.countplot(x="RENAL_CHRONIC", data=df, ax=axs[0], color="skyblue")
axs[0].set_title("Count of Renal chronic")
axs[0].bar_label(axs[0].containers[0])
sns.countplot(x="PNEUMONIA", data=df, ax=axs[1], color="salmon")
axs[1].set_title("Count of Pneumonia")
axs[1].bar_label(axs[1].containers[0])
plt.tight_layout()
plt.show()
```

```python
#Biểu đồ đếm số lượng cột CARDIOVASCULAR và HIPERTENSION
fig, axs = plt.subplots(ncols=2, figsize=(15, 7))

sns.countplot(x="CARDIOVASCULAR", data=df, ax=axs[0], color="skyblue")
axs[0].set_title("Count of Cardiovascular")
axs[0].bar_label(axs[0].containers[0])
sns.countplot(x="HIPERTENSION", data=df, ax=axs[1], color="salmon")
axs[1].set_title("Count of Hipertension")
axs[1].bar_label(axs[1].containers[0])
plt.tight_layout()
plt.show()
```

```python
#Biểu đồ đếm số lượng cột OBESITY và TOBACCO
fig, axs = plt.subplots(ncols=2, figsize=(15, 7))

sns.countplot(x="OBESITY", data=df, ax=axs[0], color="skyblue")
axs[0].set_title("Count of Obesity")
axs[0].bar_label(axs[0].containers[0])
sns.countplot(x="TOBACCO", data=df, ax=axs[1], color="salmon")
axs[1].set_title("Count of Tobacco")
axs[1].bar_label(axs[1].containers[0])
plt.tight_layout()
plt.show()
```

```python
#Biểu đồ đếm số lượng cột DIABETES và OTHER_DISEASE
fig, axs = plt.subplots(ncols=2, figsize=(15, 7))

sns.countplot(x="DIABETES", data=df, ax=axs[0], color="skyblue")
axs[0].set_title("Count of Diabetes")
axs[0].bar_label(axs[0].containers[0])
sns.countplot(x="OTHER_DISEASE", data=df, ax=axs[1], color="salmon")
axs[1].set_title("Count of Other disease")
axs[1].bar_label(axs[1].containers[0])
plt.tight_layout()
plt.show()
```
